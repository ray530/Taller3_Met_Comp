```python
import numpy as np 
import scipy as sc
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import tensorflow.keras as kr
import tensorflow as tf
import tqdm
```

## Correr solo cuando se quieran sacar los datos

```python
num=10000
A=np.random.randint(2, size=num).astype(float)
B=np.random.randint(2, size=num).astype(float)

out=np.zeros(num)
out[np.where(A==B)]=1.

df=pd.DataFrame()
df["A"]=A
df["B"]=B
df["Out"]=out

df.to_csv("Set1.csv")
```

## Punto a 

```python
data = pd.read_csv("Set1.csv")

XTrain = np.array(data[['A','B']])
# Target
YTrain = np.array(data.Out)
YTrain = YTrain[:,np.newaxis]

print(XTrain.shape, YTrain.shape)
```

```python
NN = [2,2,2,1]
lr = 0.01
```

```python
# Estructura del Modelo
model = kr.Sequential()
```

```python
# Capa 1
model.add(kr.layers.Dense(NN[1],activation='relu'))
# Capa 2 
model.add(kr.layers.Dense(NN[2],activation='relu'))
# Capa 3 
model.add(kr.layers.Dense(NN[3],activation='sigmoid'))
```

```python
# Compilar el Modelo
model.compile(loss='mse',optimizer=kr.optimizers.SGD(lr=lr),metrics=['acc'])
```

```python
# Entrenamiento
```

```python
def Train(epochs):
    
    history = model.fit(XTrain,YTrain,epochs=epochs,batch_size=XTrain.shape[0],verbose=0)
    
    return history
```

```python
for i in range(700):
    history = Train(500)
    loss = history.history['loss'][-1]
    print(i,'--- Entrenando Red --- loss:', loss)
    if loss < 0.01:
        print('Red Entrenada')
        break
```
